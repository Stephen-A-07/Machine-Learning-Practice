```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
print(train_data.shape)
print(test_data.shape)
```

```python
temp = train_data.isnull().sum() / len(train_data)
cols_to_drop = temp.loc[temp.values > 0.15].index.tolist()
print(cols_to_drop)
```

```python
for col in cols_to_drop:
    if col in train_data.columns.tolist():
        train_data.drop(col, axis = 1, inplace = True)
```

```python
train_data.dropna(inplace = True)
print(train_data.shape)
```

```python
with open('data_description.txt', 'r') as file:
    data_instructions_text = file.read()
```

```python
import re
from collections import defaultdict

data_instruction_dict = defaultdict(dict)
for col in train_data.columns:
    pattern = rf"{re.escape(col)}:\s*(.*?)(?=\n[A-Za-z0-9_]+:\s|\Z)"
    match = re.search(pattern, data_instructions_text, re.DOTALL)
    if match:
        data_instruction_dict[col] = match.group(1).strip()

parsed_dict = {}
for feature, text in data_instruction_dict.items():
    lines = [line.strip() for line in text.splitlines() if line.strip()]
    description = lines[0]
    values = {}
    for line in lines[1:]:
        parts = re.split(r'\s+', line, maxsplit=1)
        if len(parts) == 2:
            code, meaning = parts
            values[code] = meaning

    parsed_dict[feature] = {
        "description": description,
        "values": values
    }
```

```python
from pprint import pprint
pprint(parsed_dict)
```

```python
pd.set_option('display.max_columns', 100)
train_data.head(2)
```

```python
categorical_cols = dict()
numerical_cols = dict()
for key in parsed_dict:
    if len(parsed_dict[key]['values']) > 1:
        categorical_cols[key] = train_data[key].nunique()
    else:
        numerical_cols[key] = train_data[key].nunique()
```

```python
print(categorical_cols)
print(numerical_cols)
```

```python
train_data_cat = train_data[categorical_cols.keys()]
train_data_num = train_data[numerical_cols.keys()]
```
